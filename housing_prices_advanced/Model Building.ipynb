```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import warnings
warnings.filterwarnings('ignore')
```

```python
scaler = StandardScaler()

X = train.drop('SalePrice', axis=1)
y = np.log(train['SalePrice'])
```

```python
def model_build(model, X, y):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
    
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    if model.lower() == 'linear':
        
        ml_model = LinearRegression()
        ml_model.fit(X_train_scaled, y_train)
        
    elif model.lower() == 'knn':
        
        ml_model = KNeighborsRegressor()
        params = {'n_neighbors': list(range(50))}
        grid_search = GridSearchCV(ml_model, params, scoring='neg_mean_squared_error', cv=10)
        grid_search.fit(X_train_scaled, y_train)
        
        ml_model = KNeighborsRegressor(n_neighbors=grid_search.best_params_['n_neighbors'])
        ml_model.fit(X_train_scaled, y_train)
        
    elif model.lower() == 'ridge':
        ml_model = Ridge()
        params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]}
        grid_search = GridSearchCV(ml_model, params, scoring='neg_mean_squared_error', cv=10)
        grid_search.fit(X_train_scaled, y_train)
        
        ml_model = Ridge(alpha=grid_search.best_params_['alpha'])
        ml_model.fit(X_train_scaled, y_train)
        
    elif model.lower() == 'lasso':
        ml_model = Lasso()
        params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]}
        grid_search = GridSearchCV(ml_model, params, scoring='neg_mean_squared_error', cv=10)
        grid_search.fit(X_train_scaled, y_train)

        ml_model = Lasso(alpha=grid_search.best_params_['alpha'])
        ml_model.fit(X_train_scaled, y_train)
    
    elif model.lower() == 'xgboost':
        ml_model = XGBRegressor()
        params = {'learning_rate': [0.001, 0.01, 0.1, 0.4],
                 'max_depth': list(range(1, 5)),
                 'max_leaf_nodes': list(range(1, 5)),
                 'subsample': [0.25, 0.5, 0.75, 1]}
        grid_search = GridSearchCV(ml_model, params, scoring='neg_mean_squared_error', cv=5)
        grid_search.fit(X_train_scaled, y_train)
        
        ml_model = XGBRegressor(learning_rate=grid_search.best_params_['learning_rate'],
                               max_depth=grid_search.best_params_['max_depth'],
                               max_leaf_nodes=grid_search.best_params_['max_leaf_nodes'],
                               subsample=grid_search.best_params_['subsample'])
        ml_model.fit(X_train_scaled, y_train)
        
   
    model_pred = ml_model.predict(X_test_scaled)
        
    mse = mean_squared_error(y_test, model_pred)
    print('Mean Squared Error: %0.3f' % (mse))
    rmse = np.sqrt(mean_squared_error(y_test, model_pred))
    print('Root Mean Squared Error: %0.3f' % (rmse))
    mae = mean_absolute_error(y_test, model_pred)
    print('Mean Absolute Error: %0.3f' % (mae))
    print('\n')
    r_squared = ml_model.score(X_test_scaled, y_test)
    print('R-Squared: %0.2f' % (r_squared))
    adj_r_squared = 1 - ((1-r_squared) * len(X) / (len(X) - len(X.columns) - 1))
    print('Adjusted R-Squared: %0.2f' % (adj_r_squared))
    print('\n')
    train_data_r_squared = ml_model.score(X_train_scaled, y_train)
    print('Training Data R-Squared: %0.2f' % (train_data_r_squared))
    
    df = pd.DataFrame([[model, mse, rmse, mae, r_squared, adj_r_squared, train_data_r_squared]], 
                      columns=['Model', 'MSE', 'RMSE', 'MAE', 'R-Squared', 'Adj. R-Squared', 'Train R-Squared'])
    return df
```

## 1. Linear Regression

```python
comparison_df = pd.DataFrame(columns=['Model', 'MSE', 'RMSE', 'MAE', 
                                      'R-Squared', 'Adj. R-Squared', 'Train R-Squared'])

linear_model = model_build('Linear', X, y)
```
